```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://forum.12ozprophet.com/topic/47035-dont-call-it-frisco/?page=3233'
```

```python
pages = list(range(3228, 3232))
```

```python
base_url = "https://forum.12ozprophet.com/topic/47035-dont-call-it-frisco/?page="
images = set()
for page in pages:
    url = base_url+str(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    results = soup.find_all('img', class_= "ipsImage ipsImage_thumbnailed") 
    results = ["http:"+i["src"] for i in results] 
    for result in results:
        images.add(result)
        print(images)
```
